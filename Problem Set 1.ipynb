    internal class Q3
    {
        public double trianglearea(int a, int b, int c)
        {
            double s;
            s = (a + b + c) / 2;
            double area = Math.Sqrt(s * (s - a) * (s - b) * (s - c));
            return area;
        }
        public static void main(String[] args)
        {
            Q3 q3 = new Q3();
            double result = q3.trianglearea(2, 2, 2);
            Console.WriteLine(result);
        }
    }
